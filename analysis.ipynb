### Analysis

**Nodes:** Each node in the tree represents a decision point, where the data is split based on the value of a certain feature. The top node, called the root, represents the entire dataset.

**Splits/Criteria:** At each node, you'll see a condition or criterion based on one of the features. This criterion is used to split the data. For example, a node might say "Feature X <= 2.35". This means that the dataset is split at this node based on whether the value of Feature X is less than or equal to 2.35 or greater.

**Branches:** The lines coming out of a node represent the possible outcomes of the decision made at that node. In a binary tree (like most decision trees), there are typically two branches from each node.

**Leaf Nodes:** The bottom nodes of the tree, where no further splitting occurs, are called leaves. Each leaf node represents a final decision or prediction made by the tree. In a regression tree (like the one we built), this would typically be the predicted value for the target variable given the conditions leading to that leaf.

**Gini Impurity/Value:** Inside each node, you'll often see metrics like Gini impurity (for classification trees) or the mean/average value of the target variable (for regression trees). These give you an idea of how "pure" each node is - in other words, how similar or different the data points in that node are from each other in terms of the target variable.

**Samples:** This tells you how many data points from the training set fall into each node. This is useful for understanding the distribution of the data at each stage of the decision-making process.

**Colors:** The colors in the nodes usually indicate the predominant class (in classification trees) or the average value of the target variable (in regression trees). Warmer colors might represent higher values and cooler colors lower values.

In summary, the decision tree visualization helps to understand which features are important in predicting the target variable and how different values of these features influence the prediction. It's a step-by-step guide to how the model makes its predictions based on the input features.

**Justification of the Decision Tree Algorithm**
Interpretability: Decision trees are highly interpretable models. They provide clear insights into how input features affect the predictions, making them suitable for scenarios where understanding the decision-making process is as important as the accuracy of predictions.

**Non-Linear Relationships:** Decision trees do not require any assumption about the linearity between the target and the features. This makes them flexible in capturing complex, non-linear relationships that might exist in the data.

**Handling Different Types of Features:** They can handle both numerical and categorical variables effectively, although our model primarily dealt with numerical data.

**Model Building and Evaluation**
Initial Model: The first model was built using the default parameters of DecisionTreeRegressor in scikit-learn. This model achieved a very high R² score of 0.998. However, such a high score raised concerns about overfitting.<br>

**Revised Model:** To address potential overfitting, the model was revised with the following constraints:<br>

**max_depth=5:** Limits the depth of the tree to prevent the model from capturing noise in the data.<br>
**min_samples_split=20:** Requires at least 20 samples to split an internal node, avoiding overly specific splits.<br>
**min_samples_leaf=10:** Each leaf node must have at least 10 samples, ensuring that the model's predictions are based on a reasonable number of data points.<br>
**max_features='sqrt':** The number of features to consider when looking for the best split is set to the square root of the total number of features. This introduces randomness in feature selection, adding a layer of regularization.<br>
**Evaluation Metrics:** The model was evaluated using Mean Absolute Error (MAE), Mean Squared Error (MSE), and the Coefficient of Determination (R² score). The revised model had an R² score of 0.978, indicating a high but more realistic predictive performance compared to the initial model.<br>
